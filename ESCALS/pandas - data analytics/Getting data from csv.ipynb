### assign() to create a column in df
- this will take a copy and not view of the df and as a result it will not change
- we can use copy to make it reflect the df all the time

```python
dd.timestamp = pd.to_datetime(dd.timestamp)
# dd.dtypes
weekdays = dd.timestamp.dt.weekday
weekdays
dd=dd.assign(weekdays=weekdays)

weekday_names = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
week_dict = {key : weekday_names[key] for key in range(7)}

def return_day_of_week(idx):
    return week_dict[idx]

dd.weekdays = dd.weekdays.apply(return_day_of_week)
# return_day_of_week(0)
dd.head()
```

```python
weekday_counts = dd.groupby("weekdays").count()
weekday_counts = weekday_counts.loc[weekday_names]
```

```python
weekday_counts.plot(kind="bar")
```
